```python
import imp 
import numpy as np 
import geneticalgorithm as ga
import matplotlib.pyplot as plt
imp.reload(ga)
```

```python
f = lambda x: (1-x[0])**2+50*(x[1]-x[0]**2)**2 #función de prueba Rosenbrock
```

```python
I = [-3, 3]
x = np.linspace(I[0],I[1],num=100)
x,y = np.meshgrid(x,x)
p = np.array([x,y])
vals = f(p)
```

```python
results = ga.GA(f, 5e-3, 7000, 2, 20, I, npop = 220, nkeep = 30)
xtemp, ytemp = results[2].T

print(results[1])
print("\n")
print(results[0])
print(results[0][0]**2)

# La función Rosebrock tiene mínimos globales en puntos que 
# tienen la forma (a, a^2) se puede ver que esto se cumple aproximadamente 
# para los puntos encontrados con el algoritmo genético.
```

```python
plt.figure(figsize=(8,5), dpi=120)
plt.scatter(x,y,c=vals,cmap='jet')
plt.plot(xtemp,ytemp,color='white', marker='.')
plt.colorbar()
plt.show() 
```
